**Introdução
O que é Análise Exploratória de Dados (EDA)?**

Análise Exploratória de Dados, ou EDA (Exploratory Data Analysis), é o processo de investigar e compreender conjuntos de dados, resumindo suas principais características por meio de estatísticas descritivas e visualizações gráficas. Essa etapa é fundamental para revelar padrões, identificar anomalias, testar hipóteses e validar suposições antes de construir modelos preditivos ou aplicar técnicas de aprendizado de máquina. Ferramentas comuns da EDA incluem histogramas, box plots, gráficos de dispersão (scatter plots), entre outros.

**Como realizar a Análise Exploratória de Dados?**

Não existe um único método fixo para realizar a EDA, pois ela varia conforme a natureza dos dados e os objetivos da análise. Neste tutorial, vamos explorar alguns dos métodos e gráficos mais usados na prática da EDA para ajudar a extrair insights relevantes e preparar os dados para modelagem.

Em Python, a função describe() da biblioteca pandas fornece um resumo estatístico dos dados numéricos, como colunas do tipo int e float.

Esse resumo ajuda a compreender o comportamento geral da amostra, permitindo identificar:

Potenciais outliers

Erros na entrada de dados

A distribuição dos dados (se são normalmente distribuídos ou apresentam assimetria para a esquerda ou direita)

A função retorna informações como contagem, média, desvio padrão, valores mínimos e máximos, além dos quartis (25%, 50%, 75%).

**Estatísticas descritivas adicionais**

Às vezes, precisamos consultar estatísticas descritivas específicas de forma isolada ou adicionar novos parâmetros além dos fornecidos pela função describe(). Para isso, podemos usar uma abordagem complementar, calculando manualmente essas métricas e organizando-as em um resumo estatístico.

Estatísticas que podemos calcular:

**median:** Mediana das colunas.

**mode:** Moda (valor mais frequente).

**range:** Intervalo (diferença entre valor máximo e mínimo).

**cv:** Coeficiente de variação (desvio padrão dividido pela média).

**skewness:** Assimetria dos dados (skewness).

**kurtosis:** Curtose dos dados.

**sem:** Erro padrão da média (Standard Error of the Mean).

**Q1:** Primeiro quartil (25º percentil).

**Q3:** Terceiro quartil (75º percentil).

**IQR:** Intervalo interquartil (diferença entre Q3 e Q1).


```python
from scipy.stats import skew, kurtosis

# Filtrando apenas colunas numéricas
numeric_df = df.select_dtypes(include=[np.number])

# Calculando estatísticas básicas
statistics_summary = numeric_df.describe().T

# Calculando estatísticas adicionais
statistics_summary['median'] = numeric_df.median()
statistics_summary['mode'] = numeric_df.mode().iloc[0]
statistics_summary['range'] = numeric_df.max() - numeric_df.min()
statistics_summary['cv'] = statistics_summary['std'] / statistics_summary['mean']  # Coeficiente de variação
statistics_summary['skewness'] = numeric_df.apply(lambda x: skew(x.dropna()))
statistics_summary['kurtosis'] = numeric_df.apply(lambda x: kurtosis(x.dropna()))
statistics_summary['sem'] = numeric_df.sem()  # Erro padrão da média

# Calculando quartis e intervalo interquartil
quartiles = numeric_df.quantile([0.25, 0.75])
statistics_summary['Q1'] = quartiles.loc[0.25]
statistics_summary['Q3'] = quartiles.loc[0.75]
statistics_summary['IQR'] = statistics_summary['Q3'] - statistics_summary['Q1']

# Visualizando o resumo das estatísticas
statistics_summary = statistics_summary[['count', 'mean', 'std', 'median', 'mode', 'min', 'max', 'range', 'Q1', 'Q3', 'IQR', 'cv', 'skewness', 'kurtosis', 'sem']]

# Formatando todas as colunas para exibir duas casas decimais
statistics_summary = statistics_summary.applymap(lambda x: f"{x:.2f}")

# Exibindo o resumo das estatísticas
statistics_summary
```

**Abordagem utilizando IQR, para outliers superiores ao IQR.**

```python
def filter_above_iqr_for_column(df, column_name):
    # Calcular Q1, Q3 e IQR
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Limite superior
    upper_bound = Q3 + 1.5 * IQR

    # Filtrar valores acima do limite superior
    column_outliers = df[df[column_name] > upper_bound]

    return column_outliers

# Filtrar valores superiores ao IQR para a coluna 'XXXX'
outliers_up = filter_above_iqr_for_column(df, 'MSRP')

# Exibir os resultados
print("Valores superiores ao IQR na coluna 'MSRP':")
outliers_up
```

**Removendo utilizando o Intervalo Interquartil (IQR)**

Uma abordagem que considerara dados que estão acima ou abaixo do IQR como outliers.

**Removendo Ouliers maiores do que o IQR**

```python
def remove_above_iqr_for_column(df, column_name):
    # Calcular Q1, Q3 e IQR
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Limite superior
    upper_bound = Q3 + 1.5 * IQR

    # Filtrar valores abaixo do limite superior
    filtered_df = df[df[column_name] <= upper_bound]

    return filtered_df

# Remover valores acima do IQR para a coluna 'MSRP'
cleaned_df = remove_above_iqr_for_column(df, 'MSRP')

# Exibir os resultados
print("DataFrame após remover os valores acima do IQR na coluna 'MSRP':")
print(cleaned_df)
```

Explicação:
- Calcular Q1, Q3 e IQR: Calcula o primeiro quartil (Q1), o terceiro quartil (Q3) e o intervalo interquartil (IQR) para a coluna especificada.
- Definir Limite Superior:
Valores acima desse limite são considerados outliers superiores.
- Filtrar o DataFrame: Filtra o DataFrame para manter apenas as linhas onde o valor na coluna especificada é menor ou igual ao limite superior.

**Removendo outliers utilizando o Desvio Padrão**

Uma abordagem simples é considerar dados que estão a mais de 2 ou 3 desvios padrões da média como outliers.

**Removendo Ouliers maiores do que 3 desvio padrão**

```python
def remove_above_std_for_column(df, column_name):
    # Calcular média e desvio padrão
    mean = df[column_name].mean()
    std_dev = df[column_name].std()

    # Limite superior
    upper_bound = mean + 3 * std_dev

    # Filtrar valores abaixo do limite superior
    filtered_df = df[df[column_name] <= upper_bound]

    return filtered_df

# Remover valores acima do desvio padrão para a coluna 'MSRP'
cleaned_df = remove_above_std_for_column(df, 'MSRP')

# Exibir os resultados
print("DataFrame após remover os valores acima do desvio padrão na coluna 'MSRP':")
print(cleaned_df)
```

Explicação:
- Calcular Média e Desvio Padrão: A função remove_above_std_for_column calcula a média (mean) e o desvio padrão (std_dev) da coluna especificada.
- Definir Limite Superior: O limite superior é calculado para identificar valores que estão significativamente acima da média. Você pode ajustar esse fator se necessário.
- Filtrar o DataFrame: Filtra o DataFrame para manter apenas as linhas onde o valor na coluna é menor ou igual ao limite superior.

**Removendo Ouliers menores do que 3 desvio padrão**

```python
def remove_below_std_for_column(df, column_name):
    # Calcular média e desvio padrão
    mean = df[column_name].mean()
    std_dev = df[column_name].std()

    # Limite inferior
    lower_bound = mean - 3 * std_dev

    # Filtrar valores acima do limite inferior
    filtered_df = df[df[column_name] >= lower_bound]

    return filtered_df

# Remover valores abaixo de 3 desvios padrão para a coluna 'MSRP'
cleaned_df = remove_below_std_for_column(df, 'MSRP')

# Exibir os resultados
print("DataFrame após remover os valores abaixo de 3 desvios padrão na coluna 'MSRP':")
print(cleaned_df)
```

Explicação:
- Calcular Média e Desvio Padrão: Calcula a média (mean) e o desvio padrão (std_dev) da coluna especificada.
- Definir Limite Inferior: O limite inferior utiliza 3 desvios padrão para identificar valores que estão muito abaixo da média.
- Filtrar o DataFrame: Filtra o DataFrame para manter apenas as linhas onde o valor na coluna é maior ou igual ao limite inferior.

**Podemos utiliza métodos estatísticos adicionais**

Outras técnicas estatísticas, como modelos de regressão robustos ou métodos baseados em máquinas de vetores de suporte (SVM) para detecção de outliers, também podem ser usadas para identificar e remover outliers.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.svm import OneClassSVM

# Escolher a coluna para detecção de outliers
column_name = 'MSRP'

# Preparar os dados
data = df[[column_name]].dropna()  # Usar apenas a coluna de interesse e remover valores ausentes

# Normalizar os dados
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data)

# Inicializar e treinar o One-Class SVM
model = OneClassSVM(nu=0.05, kernel='rbf', gamma='auto')
model.fit(data_normalized)

# Detectar outliers
predictions = model.predict(data_normalized)

# Convertir as previsões: 1 significa normal, -1 significa anômalo (outlier)
data['outlier'] = predictions

# Filtrar o DataFrame para remover outliers
cleaned_df_SVM = df[data['outlier'] == 1]

# Exibir os resultados
print("DataFrame após remover outliers com One-Class SVM:")
print(cleaned_df_SVM)
```

**Resumo**

Escolher o método apropriado depende do contexto e dos requisitos da análise. O método do desvio padrão é simples e rápido, enquanto o IQR é mais robusto e frequentemente usado para dados não normalmente distribuídos. Boxplots são úteis para visualização e para definir limites mais informados.

Após identificar outliers com qualquer um dos métodos, você pode filtrá-los do DataFrame para melhorar a qualidade dos seus dados e a precisão da análise.
