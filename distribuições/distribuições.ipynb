**Distribuições discretas de probabilidade**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Gerar 100 dados para cada distribuição

# 1. Distribuição Uniforme Discreta
low, high = 1, 6  # intervalos para uma distribuição uniforme discreta (por exemplo, dados de um dado de 6 lados)
uniform_discrete_data = np.random.randint(low, high + 1, 100)

# 2. Distribuição Bernoulli
p = 0.5  # probabilidade de sucesso
bernoulli_data = np.random.binomial(1, p, 100)  # distribuição Bernoulli é um caso especial da Binomial com n=1

# 3. Distribuição Binomial
n, p = 10, 0.5  # número de ensaios e probabilidade de sucesso
binomial_data = np.random.binomial(n, p, 100)

# 4. Distribuição Poisson
lambda_ = 3  # taxa de eventos
poisson_data = np.random.poisson(lambda_, 100)

# Plotar histogramas

fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Histograma da Distribuição Uniforme Discreta
axs[0, 0].hist(uniform_discrete_data, bins=np.arange(low - 0.5, high + 1.5), color='skyblue', edgecolor='black')
axs[0, 0].set_title('Distribuição Uniforme Discreta')
axs[0, 0].set_xlabel('Valor')
axs[0, 0].set_ylabel('Frequência')

# Histograma da Distribuição Bernoulli
axs[0, 1].hist(bernoulli_data, bins=np.arange(-0.5, 1.5), color='salmon', edgecolor='black')
axs[0, 1].set_title('Distribuição Bernoulli')
axs[0, 1].set_xlabel('Valor')
axs[0, 1].set_ylabel('Frequência')

# Histograma da Distribuição Binomial
axs[1, 0].hist(binomial_data, bins=np.arange(-0.5, n + 1.5), color='lightgreen', edgecolor='black')
axs[1, 0].set_title('Distribuição Binomial')
axs[1, 0].set_xlabel('Valor')
axs[1, 0].set_ylabel('Frequência')

# Histograma da Distribuição Poisson
axs[1, 1].hist(poisson_data, bins=np.arange(-0.5, max(poisson_data) + 1.5), color='orange', edgecolor='black')
axs[1, 1].set_title('Distribuição Poisson')
axs[1, 1].set_xlabel('Valor')
axs[1, 1].set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```

Explicação do Código:

Distribuição Uniforme Discreta:
- Gera 100 dados inteiros entre 1 e 6 usando np.random.randint.

Distribuição Bernoulli:
- Gera 100 dados binários (0 ou 1) usando np.random.binomial com n=1.

Distribuição Binomial:
- Gera 100 dados usando np.random.binomial com número de ensaios n igual a 10 e probabilidade de sucesso p igual a 0.5.

Distribuição Poisson:
- Gera 100 dados usando np.random.poisson com parâmetro λ (taxa de eventos) igual a 3.

**Distribuições contínuas de probabilidade**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Gerar 100 dados para cada distribuição

# 1. Distribuição Normal
mu, sigma = 0, 1  # média e desvio padrão
normal_data = np.random.normal(mu, sigma, 100)

# 2. Distribuição Beta
alpha, beta = 2, 5
beta_data = np.random.beta(alpha, beta, 100)

# 3. Distribuição Gama
shape, scale = 2, 2  # forma e escala
gamma_data = np.random.gamma(shape, scale, 100)

# 4. Distribuição Log-normal
log_mean, log_std = 0, 1  # média e desvio padrão do logaritmo
log_normal_data = np.random.lognormal(log_mean, log_std, 100)

# Plotar histogramas

fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Histograma da Distribuição Normal
axs[0, 0].hist(normal_data, bins=20, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Distribuição Normal')
axs[0, 0].set_xlabel('Valor')
axs[0, 0].set_ylabel('Frequência')

# Histograma da Distribuição Beta
axs[0, 1].hist(beta_data, bins=20, color='salmon', edgecolor='black')
axs[0, 1].set_title('Distribuição Beta')
axs[0, 1].set_xlabel('Valor')
axs[0, 1].set_ylabel('Frequência')

# Histograma da Distribuição Gama
axs[1, 0].hist(gamma_data, bins=20, color='lightgreen', edgecolor='black')
axs[1, 0].set_title('Distribuição Gama')
axs[1, 0].set_xlabel('Valor')
axs[1, 0].set_ylabel('Frequência')

# Histograma da Distribuição Log-normal
axs[1, 1].hist(log_normal_data, bins=20, color='orange', edgecolor='black')
axs[1, 1].set_title('Distribuição Log-normal')
axs[1, 1].set_xlabel('Valor')
axs[1, 1].set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```

Explicação do Código:

Distribuição Normal:
- Gera 100 dados usando np.random.normal com média 0 e desvio padrão 1.

Distribuição Beta:
- Gera 100 dados usando np.random.beta com parâmetros α = 2 e β = 5.

Distribuição Gama:
- Gera 100 dados usando np.random.gamma com parâmetros de forma 2 e escala 2.

Distribuição Log-normal:
- Gera 100 dados usando np.random.lognormal com média do logaritmo 0 e desvio padrão do logaritmo 1.
